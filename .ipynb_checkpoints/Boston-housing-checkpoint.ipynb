## Постановка задачи

* загрузить данные в DataFrame
* построить график зависимости значений `medv` и `dis`
* создать отдельный DataFrame согласно условию
* сохранить новый DataFrame

Загружаем данные в DataFrame

```python
import numpy as np # импортируем библиотеку Numpy
import pandas as pd # импортируем библиотеку Pandas
import matplotlib.pyplot as plt # импортируем библиотеку Matplotlib
```

```python
boston_data = pd.read_csv('https://raw.githubusercontent.com/Andruchelli/Boston-housing/main/train.csv') # прочитаем файл
boston_data.head() # проверяем, что файл прочитался
```

Строим график зависимости значений `medv` (медианная стоимость домов, занятых владельцами, в тысячах (1000) долларов) и `dis` (среднее значение расстояний до 5 центров занятости города Бостон)

```python
x = boston_data['medv'] # задаём координату x
y = boston_data['dis'] # задаём координату y
plt.plot(x, y, 'o') # с помощью метода plot отображаем график (используем стиль отображения данных точками - 'o')
plt.xlabel('medv') # задаём подпись координаты x
plt.ylabel('dis') # # задаём подпись координаты y
```

Создаём отдельный DataFrame, содержащий только данные для домов, чья цена выше средней цены всех домов в исходных данных

```python
#mean_cost = boston_data['medv'].agg(['mean'])
mean_cost = boston_data['medv'].mean() # определяем среднее значение в столбце medv

mean_cost_data = boston_data[boston_data['medv'] > mean_cost] # создаём новый dataframe со значениями больше среднего по столбцу medv
mean_cost_data.head()
```

```python
mean_cost_data.to_csv('rich_boston.csv') # cохранияем новый DataFrame в файл rich_boston.csv
```
